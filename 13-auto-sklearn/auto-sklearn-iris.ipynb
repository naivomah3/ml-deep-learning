### What is Auto-Sklearn? 

It is a wrapper around scikit-learn, automatically searches the right machine learning algorithm for a new machine learning dataset and optimizes its hyperparameters. The following example shows the basic work of Auto-Sklearn.  

For more information, please [here](https://automl.github.io/auto-sklearn/master/index.html)

```python
from sklearn import datasets
import autosklearn.classification as classifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X, y = datasets.load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# 
automl_classifier = classifier.AutoSklearnClassifier(time_left_for_this_task=180, per_run_time_limit=40)
automl_classifier.fit(X_train, y_train)
```

```python
y_pred = automl_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(f"Model info: \n{automl_classifier.show_models()}")
```
