# Dimensionality Reduction

## **1. Principal Component Analysis**
<b>Principal Component Analysis (PCA)</b> is a statistical procedure that uses an orthogonal transformation which converts a set of correlated variables to a set of uncorrelated variables. PCA is a most widely used tool in EDA(Exploratory Data Analysis) and in machine learning for better predictive models. <br>
Moreover, PCA is an unsupervised statistical technique used to examine the interrelations among a set of variables a.k.a a general factor analysis where regression determines a line of best fitting.
<br>
<h3><b>PCA stages:</b></h3>
Given the dataset $ D ={ {x_1, x_2, ..., x_n} }$. Each $x_i$ is a $d$-dimensional vector.<br>
1. Compute the Sample Mean: $ \hat{\mu} = \frac{1}{n} \sum_{k=1}^{n} x_k $ and centralize the data: $ z_i = x_i - \hat{\mu} $ <br> 
3. Compute the Scatter Matrix: $ S = \sum_{i=1}^{n} z_i z_i^t $ <br/> 
4. Compute Eigenvectors $ e_1, e_2, ..., e_k $ corresponding to the $k-$largest Eigenvalues of the Matrix $S$ <br>
5. The desired dataset which is the closest approximation to x is $ y = E_k^t z$  such that $E_k$ is a a row matrix of $k-th$ largest Eigenvectors 

<div class="alert alert-info">
In the following we will perform PCA considering 1 and 2 Principal Components as result of the scatter matrix transformation.
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

### Load the dataset 

```python
data_iris = datasets.load_iris()
X, y = data_iris.data, data_iris.target
```

### Compute 1 and 2 Principal Components 

```python
# Centralize data using sample mean
Z = X - np.mean(X, axis=0)
# Compute the Scatter Matrix 
S = np.dot(Z.T, Z)
# Compute Eigenvalues and the corresponding  
eig_value, eig_vector = np.linalg.eig(S)
# Using 1 PC(1 Eigen-Vector of 1st largest Eigen-Value)
E1 = eig_vector[0:1,:]
# Using 2 PCs(2 Eigen-Vectors of 2 largest Eigen-Values)
E2 = eig_vector[0:2,:]    

# Compute Z projected into 1D 
Z_proj_1d = np.dot(Z, E1.T)
# Compute Z projected into 2D
Z_proj_2d = np.dot(Z, E2.T)
```

### Plot dataset using 2 PCs

```python
plt.figure(dpi=100)
plt.scatter(Z_proj_2d[:,0], Z_proj_2d[:,1], c=y)
plt.title("Projection using 2 PCs")
plt.show()
```

### Plot dataset using 1 PC

```python
# Plot dataset into 1D 
plt.figure(dpi=100)
y_1d = np.ones(Z_proj_1d.shape)
plt.scatter(Z_proj_1d[:,0], y_1d, c=y) 
plt.title('Projection using 1 PC')
plt.show()
```

### Classification using Bayesian classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
X_train_PC2, X_test_PC2, y_train, y_test = train_test_split(Z_proj_2d, y, test_size=0.3)
# Check accuracy
model = GaussianNB()
model.fit(X_train_PC2, y_train)
y_predict = model.predict(X_test_PC2)
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy: {:05.5f} %".format(accuracy))
```
