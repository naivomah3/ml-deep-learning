# Probalibilty Density Function 

## Normal distrubution $ N(\mu, \sigma^2) $

### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Initialize $\mu$ and $\sigma$

```python
mu = 0
sigma = 0.8
d = 1   # dataset dimension
```

```python
x = np.linspace(-3, 3, 600)
y = 1 / (sigma * np.sqrt(2 * np.pi)) *  np.exp( - (x - mu)**2 / (2 * sigma**2))
plt.plot(x.flatten(), y.flatten())
plt.show()
```
