# Probalibilty Density Function 

## Normal distrubution $ P(X) \sim \mathcal{N}(\mu,\,\sigma^{2})\, $
General formula of the PDF <br>
$$
\mathcal{N}(\mu,\sigma) = \frac{1}{ \sigma \sqrt{2 \pi}} e^{\left(-\frac{{\left(\mu - x \right)}^{2}}{2 \, \sigma^{2}}\right)}
$$

### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.gcf()
fig.set_size_inches(8,5)
```

### Initialize $\mu$ and $\sigma$

```python
mu = 100
sigma = 15
d = 1   # dataset dimension
```

```python
x = np.linspace(50,150,100)
y = 1 / (sigma * np.sqrt(2 * np.pi)) *  np.exp( - (x - mu)**2 / (2 * sigma**2))
plt.text(122,0.023,'$\mu = 100$',fontsize=16)
plt.text(122,0.021,'$\sigma = 15$',fontsize=16)
plt.grid(True)
plt.title('Probability Density Function')
plt.plot(x,y,color='gray')
plt.fill_between(x,y,0,color='#c0f0c0')
plt.xlabel('IQ')
plt.ylabel('Y')
plt.plot(x.flatten(), y.flatten())
plt.show()
```

## Uniform distrubution $U(a,b)$
General formula of the PDF 
$$ U(a,b) = \frac{1}{(B-A)}, \quad \text{for} \quad A \leq x \leq B; \quad 0 \quad \text{otherwise} $$

```python
a = 30
b = 50
y = []
x = [i for i in range(100)]

for i in range(100):
    # n = np.random.randint(low=1, high=100)
    if a < i < b:
        y.append(1/(b-a))
    else:
        y.append(0)    
plt.plot(x, y, linewidth=0.5, color='b')
plt.text(65,0.023,'$a = 30$',fontsize=16)
plt.text(65,0.015,'$b = 50$',fontsize=16)
plt.grid(True)
plt.title('Probability Density Function')
plt.show()
```

## Exponential Distribution 
General formula
$$
P(x) = \lambda e^{-\lambda x}, \quad \text{for} \quad x \gt 0 
$$

```python
# lb = standard deviation
lb = 1.5   
# Generate a 600 lines of data between [-3,3] 
# This interval [-3,3] will be divided into 600 parts to get the 600 lines of data
x = np.linspace(-3, 3, 600)
# print(len(x))
# x = np.random.normal(mu, sigma, 100)
y = lb * np.exp( - lb * x)
plt.grid(True)
plt.text(2,10,'$\lambda = 1.5$',fontsize=16)
plt.title('Probability Density Function')
plt.plot(x, y, linewidth=1, color='g')
plt.show()
```

## Poisson Distribution 
General formula
$$
P(X=k) = \frac{\lambda^{k} e^{-\lambda}} {k!} \quad \text{for} \quad \lambda \gt 0 
$$

```python
import math
# lambda = Expected value E(x)
def poisson(k, lb):
    return (lb**k / math.factorial(k)) * np.exp(-lb)

lb = 14
x = [xi for xi in range(100)]
y = []
for i in range(100):
    y.append(poisson(i, lb))
    
plt.grid(True)
plt.text(60,0.025,'$\lambda = 14$', fontsize=16)
plt.title('Probability Density Function')
plt.plot(x, y, linewidth=1, color='g')
plt.show()
```
